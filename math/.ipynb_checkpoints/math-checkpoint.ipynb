```python
#최대공약수
def getGCD(M, N):
    if M>N: N, M = M, N
    for num in range(M, 0, -1):
        if M % num == 0 and N % num == 0:
            return num # 최대공약수

print(getGCD(12,18))
    # for num in range(1, m+1):
    #     if M % num == 0 and N % num == 0:
    #         cdivisor_list.append(num)

```
