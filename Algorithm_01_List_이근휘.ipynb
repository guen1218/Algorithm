```python
#파이썬 리스트 연습 : 항목 접근
test_list = [1, 2, 3, 'one', 'two', 'three']
print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list :
    print (item)

#파이썬 리스트 연습 : 항목 삽입과 삭제

test_list.insert(3, 4)

print("\ninsert(3, 4)")
for item in test_list :
    print (item)

test_list.append('four')

print("\nappend('four')")
for item in test_list :
    print (item)

test_list.remove('one')

print("\nremove('one')")
for item in test_list :
    print (item)

item = test_list.pop(0)

print("\npop(0)")
for item in test_list :
    print (item)

item = test_list.pop()

print("\npop()")
for item in test_list :
    print (item)
```

```python
#리스트의 반복 구조
#주어진 데이터가 어디에 저장되어 있는지 출력
data_list = [23,45,21,14,2,67]
target = 14
index = 0

while index <= 5 :
    if data_list[index] == target :
        break
    index += 1
print (index)
```

```python
data_list = [23,45,21,14,2,67]
target = 14
for index in range(6):
    if data_list[index] == target:
        break
print (index)
```

```python
#리스트에 저장된 숫자 중 가장 큰 숫자 출력(최대값)
data_list = [123,45,10,14,2,67]

max = data_list[0]
max_index = 0
for index in range (1, 6):
    if data_list[index] > max :
        max = data_list[index]
        max_index = index
        
print (max_index, max)
```

```python
data_list = [123,45,10,14,2,67]

max = data_list[0]
for data in data_list :
    if data>max:
        max = data

print (max)
```

```python
#리스트에 저장된 숫자 중 가장 작은 숫자 출력(최소값)
data_list = [13,45,10,14,2,67]

min = data_list[0]
for data in data_list :
    if data<min:
        min = data

print (min)
```

```python
#리스트에 저장된 숫자 중 입력된 숫자보다 큰 수가 몇 개 있는지 구하기
data_list = [25,45,10,14,2,67]

num = int(input())
count = 0 
for i in data_list :
    if num < i :
        count += 1
print (count)
```

```python
#리스트에 저장된 모든 값 더하여 출력 (합)
data_list = [13,45,10,14,2,67]

sum = 0
for i in data_list :
    sum += i
print (sum)
```
