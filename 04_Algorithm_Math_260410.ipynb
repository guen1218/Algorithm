```python
def prime_number(N):
    is_prime_number = True
    if N <= 1 or N % 2 == 0:
        is_prime_number = False
    else :
        for j in range (3, N//2):
            if N%j == 0:
                is_prime_number = False
                break
    return is_prime_number
```

```python
# 정수 N을 받아서 base에 대한 진법의 수로 변환
def convent_base(N,base):
    #N을 base로 나눈 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    # N : 몫을 저장
    #new_list : 나머지를 저장하는 링크
    new_number = ''
    digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    while N!=0:
        new_number = digits[N%base] + new_number
        N //= base
    return new_number
print(convent_base(25,2))
```
