```python
#데이터 정의
stack = []

while True :
    # 메뉴를 출력
    print ('''
    ======================================
    1. Push    2. Pop    3. Peek    0.Exit
    ======================================
    ''')
    
    #입력을 받는다
    menu = int(input(">> 연산 메뉴 선택 : "))
    # print(menu)
    
    #입력에 따라 stack 연산을 수행하고
    if menu == 0 : 
        print("정수형 스택 종료") 
        break
        
        
    elif menu == 1 : #push
        data = int(input(">> [Push] 추가할 데이터 입력 : "))
        stack.append(data)
        
        
    elif menu == 2 : # pop
        if len(stack) != 0: # error
            data = stack.pop()
            print ("[Pop] 가져온 데이터 : ", data)
        else :
            print ("다시하세요라")
            continue

        
    elif menu == 3 : # peek
        if len(stack) != 0: # error
            # data = stack[len(stack)-1]
            print ("top 데이터 확인 : ", stack[len(stack)-1])
            continue
        else :
            print ("다시하세요라")
            continue
        
    else : # error
        print ("다시하세요라")
        continue
    
    #스택의 상태를 출력한다
    print ("[Stack]", stack)
    
    #위의 과정을 0. Exit을 선택할 때까지 반복
```

```python
# BOJ 10828
N = int(input())
stack=[]
for _ in range(N):
    operation = input().split()
    if operation[0] == 'push':
        stack.append(int(operation[1]))
    elif operation[0] == 'pop':
        print (stack.pop() if stack else -1)
    elif operation[0] == 'size':
        print(len(stack))
    elif operation[0] == 'empty':
        print (1 if not stack else 0)
    elif operation[0] == 'top':
        print (stack[-1] if stack else -1)
```
