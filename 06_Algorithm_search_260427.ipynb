```python
def binary_search_rec(A, key, low, high): # 함수지정
    if (low>high): # 못찾고 많이 돌아가서 로우가 하이를 넘었을 경우 (찾는 값이 없을 경우)
        return -1 # -1
    middle = (low+high) // 2 # 가운데 인덱스 구하기

    if (key == A[middle]): # 만약 가운데 인덱스가 키일 경우
        return middle # 가운데 값 내보냄
    elif (key < A[middle]): # 키가 가운데 인덱스 위치 친구보다 작을 경우
        return binary_search_rec(A, key, low, middle - 1) # 하이를 미들 - 1(미들을 또 구분할 필요는 없다) 해서 리커시브 (반을 자름)
    else : # 아니면(키가 더 클경우)
        return binary_search_rec(A, key, middle + 1, high) # 로우를 미들 +1 해서 리커시브
```

```python
def binary_search(A, key):# 함수지정
    low = 0 # 로우 첫번째칸
    high = len(A) - 1 # 하이 마지막 인덱스 번호
    while low <= high: # 로우가 하이보다 작을 경우에만 (로우가 하이보다 커지면 못찾았다는 뜻)
        middle = (low+high)//2 # 중간 인덱스값 구하기
        if A[middle] == key: # 미들이 키일 경우 리턴 미들 
            return middle
        elif A[middle] > key: # 미들이 키보다 클 경우 
            high = middle -1 # 하이를 미들 - 1 하고 와일 다시
        else: # 아니라면( 미들 < 키)
            low = middle +1 # 로우를 미들 + 1 하고 와일 다시
    return -1 # 찾는 값이 없을 경우
```
